```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('titanic.csv')

print("Missing Values:\n", data.isnull().sum())  
data.dropna(inplace=True)  

print("Duplicate Rows:", data.duplicated().sum()) 
data.drop_duplicates(inplace=True)

print("Data Types:\n", data.dtypes)  

print(data.describe())
print(data.info())

plt.figure(figsize=(10, 6))
data.select_dtypes(include=np.number).hist(bins=30, figsize=(10, 6))
plt.suptitle('Histogram of Numeric Features')
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(data=data.select_dtypes(include=np.number))  
plt.title('Boxplot of Numeric Features')
plt.xticks(rotation=90)  
plt.show()

plt.figure(figsize=(10, 6))
corr = data.select_dtypes(include=np.number).corr()  
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

print("Key Findings:")
print("- Patterns observed in distributions.")
print("- Correlations between variables.")
print("- Presence of outliers and their potential impact.")
```
